```python
fileToRead = "areadata2040UNFIXED.csv"
fileToWrite = fileToRead.replace("UNFIXED", "")

maps = [
    ["DK-West","DK1"],
    ["DK-East","DK2"],
    ["Holland","NL"],
    ["KriegersFlak","DKKF"],
    ["Bornholm","DKBO"],
    ["EnergyIsland","DKEI"],
    ["Belgium","BE"],
    ["Finland","FI"],
    ["Austria","AT"],
    ["France","FR"],
    ["Poland","PL"],
    ["Schweiz","CH"],
    ["Hungary","HU"],
    ["Italy", "IT"],
]
```

```python
line = True
read = open(fileToRead, "r", errors="ignore")
write = open(fileToWrite, "w+")

while(line):
    line = read.readline()
    for map in maps:
        line = line.replace(map[0],map[1])
    write.write(line)

write.close()
```

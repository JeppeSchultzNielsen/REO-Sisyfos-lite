```python
f = open("FirstSim.txt","r") 
header = f.readline()
splitted = header.split()
data = np.transpose(np.loadtxt("FirstSim.txt", skiprows = 1))
```

```python
def plot(to_plot: str, bunch_in_hours: int = 24, plot_from: int = 0, plot_to: int = 8600):
    ind = splitted.index(to_plot)
    data_to_use = data[ind][plot_from:plot_to]
    points = int(len(data_to_use) / bunch_in_hours)

    x_vals = []
    y_vals = []

    for i in range(points):
        x_vals.append(i*bunch_in_hours + plot_from)
        y_vals.append(sum(data_to_use[i*bunch_in_hours:(i+1)*bunch_in_hours]))

    plt.plot(x_vals,y_vals,label = to_plot)
    plt.ylabel("MWh per " + str(bunch_in_hours) + "hours")
    plt.xlabel("Hours")
    plt.legend()
```

```python
f = open("FirstSim.txt","r") 
header = f.readline()
splitted = header.split()
data = np.transpose(np.loadtxt("FirstSim.txt", skiprows = 1))

plot_from = 0
plot_to = 8758
bunch_in_hours = 1

ind = splitted.index("NOs_demand")
data_to_use = data[ind][plot_from:plot_to]

ind = splitted.index("DK1_to_NOs")
data_to_use -= data[ind][plot_from:plot_to]

ind = splitted.index("NOm_to_NOs")
data_to_use -= data[ind][plot_from:plot_to]

ind = splitted.index("NOs_to_SE3")
data_to_use += data[ind][plot_from:plot_to]

ind = splitted.index("NOs_to_NL")
data_to_use += data[ind][plot_from:plot_to]

ind = splitted.index("NOs_to_GB")
data_to_use += data[ind][plot_from:plot_to]

ind = splitted.index("NOs_to_DELU")
data_to_use += data[ind][plot_from:plot_to]

points = int(len(data_to_use) / bunch_in_hours)

x_vals = []
y_vals = []

for i in range(points):
    x_vals.append(i*bunch_in_hours + plot_from)
    y_vals.append(sum(data_to_use[i*bunch_in_hours:(i+1)*bunch_in_hours]))

plt.plot(x_vals,y_vals,label = "Total energy spent in NOs")
plt.ylabel("MWh per " + str(bunch_in_hours) + "hours")
plt.xlabel("Hours")
plt.legend()
```

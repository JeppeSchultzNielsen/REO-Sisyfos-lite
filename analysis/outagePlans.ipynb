```python
import numpy as np 
import matplotlib.pyplot as plt
from ResultsFile import ResultsFile
```

```python
var = ResultsFile("../results/2035_1985_var_[0].txt")
noVar = ResultsFile("../results/2035_1985_noVar_[0].txt")
```

```python
#plot outage plans
fig, ax = plt.subplots(figsize = [5,3])
histbins = np.linspace(0,8759,8760)

outage = noVar.getColumn("DK1_plannedOutage")

ax.plot(histbins,outage)
ax.set_xlabel("Hour")
ax.set_ylabel("Power lost to planned outage [MW]")
ax.set_title("Outage plan when not taking variation generation into account")

ax2 = ax.twinx()
outage = noVar.getColumn("DK1_demand")
ax2.plot(histbins,outage,'red',alpha=0.5)
ax2.set_ylabel("Demand [MW]")
```

```python
#plot outage plans
fig, ax = plt.subplots(figsize = [5,3])
histbins = np.linspace(0,8759,8760)

outage = var.getColumn("DK1_plannedOutage")

ax.plot(histbins,outage)
ax.set_xlabel("Hour")
ax.set_ylabel("Power lost to planned outage [MW]")
ax.set_title("Outage plan when taking variation generation into account")

ax2 = ax.twinx()
print(noVar.header)
outage = var.getColumn("DK1_demand")-var.getColumn("Wind_DK1")-var.getColumn("ICHP_DK1")-var.getColumn("Solar_DK1")
ax2.plot(histbins,outage,'red',alpha=0.5)
ax2.set_ylabel("Demand - productions of wind, solar [MW]")
```

```python
#plot outage plans
fig, ax = plt.subplots()
histbins = np.linspace(0,8759,8760)

outage = noVar.getColumn("DK1_demand")-noVar.getColumn("DK1_WL")-noVar.getColumn("DK1_WS")-noVar.getColumn("DK1_ICHP")

ax.plot(histbins,outage)
ax.set_xlabel("Hour")
ax.set_ylabel("Demand [MW]")
ax.set_title("Demand-other productions")
```

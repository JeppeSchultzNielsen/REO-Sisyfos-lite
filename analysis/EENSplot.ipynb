```python
#formålet med denne notebook er at reproducere figur 25 og 26 på side 40 af baggrundsrapporten
import numpy as np 
import matplotlib.pyplot as plt
from ResultsFile import ResultsFile
```

```python
climateYears = [1985,1987,1996,2007,2008,2009,2010,2011,2012,2013,2014,2015]
files = []
for i in range(len(climateYears)):
    for j in range(2):
        files.append(ResultsFile("../results/2035_"+str(climateYears[i])+"_"+str(j)+".txt"))
```

```python
hours = files[0].getColumn("hour")
DK1EENS = files[0].getColumn("DK1_EENS")
DK2EENS = files[0].getColumn("DK2_EENS")

for i in range(len(files)-1):
    DK1EENS += files[i+1].getColumn("DK1_EENS")
    DK2EENS += files[i+1].getColumn("DK2_EENS")

DK1EENS = DK1EENS/len(files)
DK2EENS = DK2EENS/len(files)
```

```python
hourlyEENSDK1 = np.zeros(24)
hourlyEENSDK2 = np.zeros(24)
hour = np.arange(0,24,1)
dailyEENSDK1 = np.zeros(365)
dailyEENSDK2 = np.zeros(365)
day = np.arange(0,365,1)
for i in range(len(DK1EENS)):
    currentHour = int(hours[i]%24)
    currentDay = int(hours[i]/(8760/365))
    hourlyEENSDK1[currentHour] += DK1EENS[i]
    dailyEENSDK1[currentDay] += DK1EENS[i]
    hourlyEENSDK2[currentHour] += DK2EENS[i]
    dailyEENSDK2[currentDay] += DK2EENS[i]
```

```python
plt.bar(hour,hourlyEENSDK1)
plt.xlabel("Time på døgnet")
plt.ylabel("EENS i DK1 over et helt år")
```

```python
plt.bar(hour,hourlyEENSDK2)
plt.xlabel("Time på døgnet")
plt.ylabel("EENS i DK2 over et helt år")
```

```python
plt.bar(day,dailyEENSDK1)
plt.xlabel("Dag på året")
plt.ylabel("EENS i DK1 over et helt år")
```

```python
plt.bar(day,dailyEENSDK2)
plt.xlabel("Dag på året")
plt.ylabel("EENS i DK2 over et helt år")
```
